```python
# taking care of missing values
train['runtime'] = train['runtime'].fillna(method='ffill')
```

```python
set(train.status)
```

```python
statusMap = {'Released':0,'Rumored':1}
```

```python
train.replace({'status': statusMap}, inplace = True)
```

```python
#train test split
X = train[['runtime', 'budget','popularity','status']]
y = train.revenue
#splitting the data into training and validation to check validity of the model

from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.1, random_state=1)
```

```python
#Linear Model
from sklearn.linear_model import LinearRegression
def rmsle(y,y0): return np.sqrt(np.mean(np.square(np.log1p(y)-np.log1p(y0)))) 
#taking log to decrease the value given in the below block
#It is the Root Mean Squared Error of the log-transformed predicted and log-transformed actual values. 
#RMSLE adds 1 to both actual and predicted values before taking the natural logarithm to 
#avoid taking the natural log of possible 0 (zero) values
#root mean value square
#xi = current value
#xiGap = mean value
# (xi - xiGap)^2 
reg = LinearRegression()
lin_model = reg.fit(X_train, y_train)
y_pred = reg.predict(X_val)
print('RMSLE score for linear model is {}'.format(rmsle(y_val, y_pred)))
```

```python
#other way we are trying
from sklearn.metrics import mean_squared_error
met = mean_squared_error(y_val, y_pred, squared=False)
print(met)
```

```python
from sklearn.metrics import r2_score

r2_res = r2_score(y_val,y_pred)
print(r2_res)
```
